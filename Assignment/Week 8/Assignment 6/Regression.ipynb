

### # Function to convert gender to binary

```python
def score_to_numeric(x):
    if x=='female':
        return 1
    if x=='male':
        return 0
```

```python
df['gender_val'] = df['gender'].apply(score_to_numeric)
df.tail()
```

```python
result = sm.ols(
    formula='grade ~ age + exercise + hours + gender_val',
    data=df).fit()
result.summary()
```

**ANS:** No change to the adjusted r-squared
